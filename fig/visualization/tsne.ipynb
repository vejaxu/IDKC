```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat

# Step 1: Load the MAT file
data_dict = loadmat('../../../data/w1Gaussians.mat')  # 替换为您的文件名

# Step 2: Extract data and class labels
data = data_dict['data']  # 假设数据存储在 'data' 变量中，确保它是二维的
class_labels = data_dict['class'].flatten()  # 假设类别存储在 'class' 变量中

# Step 3: Plot the results
plt.figure(figsize=(8, 6))
scatter = plt.scatter(data[:, 0], data[:, 1], c=class_labels, cmap='viridis', alpha=0.7)
plt.title('w1Gaussians dataset Visualization')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.colorbar(scatter, label='Class Labels')  # 添加颜色条
plt.grid(True)
plt.savefig("w1Gaussians dataset visualization.png", dpi=300)  # 保存图像
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from scipy.io import loadmat

# Step 1: Load the MAT file
data_dict = loadmat('../../../data/w2000Gaussians.mat')  # 替换为您的文件名

# Step 2: Extract data and class labels
data = data_dict['data']  # 假设数据存储在 'data' 变量中
class_labels = data_dict['class'].flatten()  # 假设类别存储在 'class' 变量中

# Step 3: Perform t-SNE
tsne = TSNE(n_components=2, random_state=42)  # 设置随机状态以确保可重复性
Y = tsne.fit_transform(data)

# Step 4: Plot the results
plt.figure(figsize=(8, 6))
scatter = plt.scatter(Y[:, 0], Y[:, 1], c=class_labels, cmap='viridis', alpha=0.7)
plt.title('t-SNE Visualization')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.colorbar(scatter, label='Class Labels')  # 添加颜色条
plt.grid(True)
plt.savefig("w2000Gaussians dataset visualization tsne.png", dpi=300)  # 保存图像
plt.show()
```
